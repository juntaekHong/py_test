```python
#sql_2="""
#    SELECT *FROM sales GROUP BY Country
#    """

sql_2="""
    SELECT 
    Country, 
    SUM(`Units Sold`) 
    AS 'units sold' 
    FROM sales 
    GROUP BY Country
    ORDER BY `units sold` DESC
    
    """

#order  by 오름차순 desc 내림차순
#공백같은 경우에 백쿼트를 넣으면 공백도 문자로 인식하여 나옴 -백쿼트 1옆에 있는 키보드
#AS를 하면서 SUM('Units Sold')에서 units sold부분으로 바뀜   
result_2=_db.executeAll(sql_2)
dt_2 =pd.DataFrame(result_2)
dt_2.head()
#그룹값들이 묶여서 제일 처음값들만 나옴
```

```python
sql_2="""
    SELECT 
    Country, 
    AVG(`Units Sold`) 
    AS 'units sold' 
    FROM sales 
    GROUP BY Country
    ORDER BY `units sold` DESC
    
    """
result_2=_db.__excuteAll__(sql_2)
dt_2 =pd.DataFrame(result_2)
dt_2.head()
```

```python
#SALES DATA
#1.그룹화 국가
#2.출력 -> COUNTRY,CONUNTRY 개수 구하는 DATAFRAME
#3.국가 개수를 기준으로 내림차순정렬
sql_3="""
    SELECT 
    Country, 
    COUNT(`Country`) 
    AS 'Con_con' 
    FROM sales 
    GROUP BY Country
    ORDER BY `Con_con` 
    """
result_3=_db.__excuteAll__(sql_3)
dt_3 =pd.DataFrame(result_3)
dt_3
```
