```python
#2000보다 크거나 같은 데이터만 출력 $gte: 2450는 포함시킴

result=db.emp.find(
    {
    "SAL":{
        "$gt":2000
        }
    }
    )
for list in result:
    print(list)
# SELECT * FROM emp WHERE SAL>2000
 #SELECT * FROM emp WHERE SAL>=2500
```

```python
#2450보다  작은 데이터만 출력 #여기서도 lt에 e를 붙여주면 
result=db.emp.find(
    {
    "SAL":{
        "$lt":2000
        }
    }
    )
for list in result:
    print(list)

# SELECT * FROM emp WHERE SAL<2450
```

```python
#SELECT * FROM emp WHERE ENAME= SMITH OR ENAME= ALLEN
result = db.emp.find(
    {
        "$or" : [
            {
                "ENAME" : "SMITH"    
            },
            {
                "ENAME" : "ALLEN"    
            }
            ]
    }
)
for list in result:
    print(list)
```

```python
#SELECT * FROM emp WHERE ENAME= SMITH OR ENAME= ALLEN
result = db.emp.find(
    {
        "$and" : [
            {
                "ENAME" : "SMITH"    
            },
            {
                "JOB" : "CLERK"    
            }
            ]
    }
)


for list in result:
    print(list)
```

```python
#SELECT * FROM emp WHERE ENAME =SMITH AND JOB = CLERK #and 따로 안써줘도 됨
result = db.emp.find(
    {
        "ENAME" : "SMITH",
        "JOB" : "CLERK"  
    }
)


for list in result:
    print(list)
```

```python
#SELECT * FROM emp WHERE SAL >= 2000 AND SAL<4000
result_4 = db.emp.find(
    {
        "SAL":{"$gte": 2000,"$lt": 4000}
        },
     {
         "_id": 0 
      }
)


pd.DataFrame(result_4)
```

```python
#SELECT * FROM emp WHERE ENAME LIKE %%SVG
result_5 = db.emp.find(
    {
        "ENAME" : {
            "$regex" : "M"
    }}
)
pd.DataFrame(result_5)
#S s를포함하는 문자
#^S s로 시작하는 문자
#S$  S로 끝나는 문자 
```
