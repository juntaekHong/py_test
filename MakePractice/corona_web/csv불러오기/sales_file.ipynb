```python
import pandas as pd
```

```python
sales_df = pd.read_csv('./csv파일/Sales Records.csv')
sales_df.columns
```

```python
##지금부터 해야할것
#1. 칼럽삭제(ship date, order id, sales channel, Total revenue) -> 원본데이터 유지
#2. 컬럼명 수정 --> 원본데이터 유지X 
#3. Country기준으로 sort
#4. index값을 초기화
sales_df_2=sales_df.drop(["Ship Date","Order ID","Sales Channel","Total Revenue"],axis=1)
sales_df_2.columns["국가"]
sales_df_2.columns=["대륙","국가","상품타입","배송순위"
                   ,"주문날짜","판매개수","단가","원가","총비용","총이윤"]
sales_df_2.shape #10000행 10열
```

```python
#sales_df_2.sort_values("국가",inplace=True)
sales_df_2=sales_df_2.sort_values(["국가"])

sales_df_2.reset_index(drop=True,inplace=True)

sales_df_2
```

```python
#변수['컬럼명'].value_counts() 컬럼에 속하는 값들의 개수
sales_df_2['대륙'].value_counts
```

```python
sales_df_2['국가'].value_counts()
# 컬럼에 속하는 국가의 개수
```

```python
#총이윤을 astype으로 한다면 총이윤 소수를 지울 수 있다.
sales_df_2['총이윤']=sales_df_2['총이윤'].astype(int)
sales_df_2
```

```python
#round쓰고 astype을 쓰면 .0까지 지워짐.
sales_df_2["총비용"].round(0).astype(int)
sales_df_2["총비용"]=sales_df_2["총비용"].round(0).astype(int)
sales_df_2
```

```python
#원가에서 소수점없애기
sales_df_2['원가']=sales_df_2['원가'].round(1).astype(int)
sales_df_2
```

```python
sales_df_2['판매개수'].round(-1).astype(int)
sales_df_2
```

```python
sales_df_2['주문날짜']=pd.to_datetime(sales_df_2['주문날짜'],format="%M/%d/%Y")
```

```python
import numpy as np

s1 = pd.Series([1,np.nan,3,4,5])
s2 = pd.Series([1,2, np.nan,4,5])
s3 = pd.Series([1,2,3,np.nan,5])
df= pd.DataFrame({'s1':s1,'s2':s2,'s3':s3})
df
```

```python
df["s1"].isna() 
#NaN값을 True변환, Nan을 fAlse로 변환
```

```python
df.isna()
```

```python
df.isna().sum()
#True 값의 개수,NaN값의 개수를 알 수 있음.
```

```python
df.isna().sum(axis=1)
#행에서 1값을 할 수 있음.
```

```python
df.dropna(axis=0)
#dropna 결측값을 제거하는 함수.
```

```python
df.dropna(axis=1)
#s1,s2,s3에 인데스값만 나옴, 왜냐하면 1값이 다 나오니깐
```

```python
df['s1']=df['s1'] +1
df
```

```python
_list = [5,4,3,2,1]
df['s2']=df['s2'] + _list
df
```

```python
_tuple = (1,2,3,4,5)
df['s3']=df['s3'] + _tuple
df
#series,tuple,list로 만들어서 사칙연산할 수 있음.
```

```python
#행렬의 곱이나 합을 구하려면 행과 곱하려는 렬의 개수가 같아야하므로
#만들어주는 것.
_list = [2,1,2] #@,.dot을 통해서 얻을 수 있음.
#dot_result = df@_list
dot_result = df.dot(_list)
dot_result
```
