```python
### class 하나 생성
1. 인자 값으로 url(주소)받아오기 -> csv파일주소
2. 함수를 하나 생성-> url에 있는 csv파일을 읽어오고, 'Country'기준으로 sort, 
3.인덱스 값을 초기화
4.함수를 하나 생성해서 -> 인자값을 리스트로 받는다. -> 이 매개변수를 기준으로 컬럼삭제
```

```python
import pandas as pd
```

```python
class Class_1():
    def __init__(self,r):
        self.url=r

    def csv_read(self):
        self.read=pd.read_csv(self.url)
        self.read.sort_values('Country',inplace=True)
        self.read.reset_index(drop =True ,inplace=True)
        return self.read
    
    def make__(self,input_a):
        self.read.drop(input_a,axis=1,inplace=True)
        
        return self.read
```

```python
a=Class_1("./Sales Records.csv")
a.csv_read()

```

```python
a.make__(["Item Type","Order Date"])
```

```python
class Class_1():###디폴트 값 만들어주기
    def __init__(self,r):
        self.url=r

    def csv_read(self,input_sort="Country",input_inplace=True):#변수 더 받아서하기
        self.read=pd.read_csv(self.url)
        self.read.sort_values(input_sort,inplace=input_inplace)
        self.read.reset_index(drop =True ,inplace=input_inplace)
        return self.read
    
    def make__(self,input_list = []):
        self.read.drop(input_list,axis=1,inplace=True)
        
        return self.read
    
    def remove_2(self, input_s_column, input_e_column):
        self.read.drop(self.read.loc[:, input_s_column: input_e_column], axis=1)
        return self.read
```

```python
a=Class_1("./Sales Records.csv")
a.csv_read()
#a.make__()
a.remove_2("Item Type","Total Cost")
```
