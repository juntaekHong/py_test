```python
#1.emp table과 dept table을 left join조건 dept가 같은 경우
#2.출력 컬럼은 ENAME,JOB,SAL,DEPTNO,DNAME,LOC
#3.ENAME은 그대로 출력
#4.JOB은 각각 한글명으로 변경
#5.SAL은 2000은 기준으로 2450 이상이면 HIGH미만이면 LOW
#6.DEPTNO은 DEPTNO의 값과 dept table의 DNAME값을 더해서 표시()
#7.모든 컬럼의 이름은 한글로 변경
```

```python
sql_18="""
        SELECT 
        emp.ENAME AS '사원명',
        CASE
            WHEN emp.JOB ="ANALYST" THEN "분석가"
            WHEN emp.JOB ="CLERK" THEN "점원"
            WHEN emp.JOB ="MANAGER" THEN "관리자"
            WHEN emp.JOB ="PRESIDENT" THEN "사장"
            WHEN emp.JOB ="SALESMAN" THEN "판매사원"
        END AS '직업',
        IF(emp.SAL >= 2450, 'HIGH', 'LOW') AS '연봉',
        concat(emp.DEPTNO,'-',dept.DNAME) AS '부서번호-부서명',
        dept.LOC AS '위치' 
        FROM emp LEFT JOIN dept 
        on emp.DEPTNO = dept.DEPTNO
        ORDER BY `사원명` ASC
        """
result_18=_dt.executeAll(sql_18)
df_18 =pd.DataFrame(result_18)
df_18  
#스트링안에 따옴표를 쓰려고 평소에 백쿼트를 넣지만 SQL문에서는 스트링이 안먹힐때 백쿼트를 적용하면 될때가 있다.
```

```python
###부서의 이름이 'sales'가 아닌 소속 사원들의 리스트
###부서의 위치가 'NEW YORK','DALLAS'가 아닌 소속 사원들의 리스트를 출력하라
```

```python
sql_23="""
        SELECT 
        ENAME 
        FROM emp
        WHERE DEPTNO =
        ANY(
         SELECT DEPTNO FROM dept
         WHERE NOT LOC = 'NEW YORK' AND LOC = 'DALLAS'
        )
        """
#컬럼에도 where사용가능
df_23 = _dt.executeAll(sql_23)
pd.DataFrame(df_23)
# =ANY 대신해서 IN을 쓸 수 있다. 
```
