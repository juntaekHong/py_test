```python
corona_df_2 =corona_df.sort_values("등록일시") #ascending 오름차순 descending 내림차순
corona_df_2
```

```python
corona_df_2["일일 확진자"] = ((corona_df_2["확진자"]) - corona_df_2["확진자"].shift()).fillna(0)
corona_df_2
```

```python
corona_df_2["일일 사망자"] = ((corona_df_2["사망자"]) - corona_df_2["사망자"].shift()).fillna(0) #fillna값은 처음값이 나올 수가 없기 때문에 nan값을 대체하는 것
corona_df_2
```

```python
corona_df_2["일일 사망자"] =corona_df_2["사망자"].diff().fillna(0) #diff 다음행과 차이 
corona_df_2
```

```python
corona_df_2["일일 사망자"].value_counts(ascending=True)
```

```python
#코로나 누적확진률이 3.5이상이면 High,3.5미만이면 Low칼럼을 추가
corona_df["level"]=corona_df["누적확진률"].apply(lambda x : "High" if ( x >= 3.5) else "Low")
corona_df
```

```python
corona_df_3["등록일시"] = pd.to_datetime(corona_df_3["등록일시"],format="%y-%m-%d")

corona_df_3["요일"] = corona_df_3["등록일시"].dt.weekday
corona_df_3["등록일시"]=pd.to_datetime(corona_df["등록일시"]).dt.normalize()
corona_df_3["등록일시"]

corona_df_3.head()
```

```python
def day(x):
    if x==0:
        return "월"
    elif x ==1: 
        return "화"
    elif x ==2:  
        return "수"
    elif x ==3 : 
        return "목"
    elif x ==4: 
        return "금"
    elif x ==5 : 
        return "토"
    elif x ==6 :
        return "일"
    else :
        return 

corona_df_3["요일"]=corona_df_3["요일"].apply(day)
corona_df_3
```

```python

def week_change_2(x):
        week=["월","화","수","목","금","토","일"]
        return week[x]

corona_df_3["요일"]=corona_df_3["요일"].apply(week_change_2)
corona_df_3
```
