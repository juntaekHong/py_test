```python
#컬럼별로 확인 가능
corona_df['누적확진률'].describe()
```

```python
corona_df_2=corona_df.copy()
corona_df_2 = corona_df_2.sort_values('등록일시')
corona_df_2["일일사망자"]=(corona_df_2["사망자"]-corona_df_2["사망자"].shift()).fillna(0)
corona_df_2["일일확진자"]=corona_df_2['확진자'].diff().fillna(0)
corona_df_2
```

```python
# shift말고 다른 함수 쓰기 !! diff! 음수 양수 상관없이 차이 구해줌 !
# corona_df_2["일일 사망자"] = corona_df_2["사망자"].diff().fillna(0)
# corona_df_2
```

```python
#축 안넣으면 오류남 axis=1 꼭 해줘야 함 !!
corona_df_3=corona_df_2.drop(["인덱스","게시글번호","기준시간","기준일","수정일시"],axis=1)
corona_df_3
```

```python
#corona_df_3.reset_index(drop=True)
corona_df_3.reset_index(drop=True,inplace=True)
```
